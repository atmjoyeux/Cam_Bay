```python
#!/usr/bin/env python
# coding: utf-8

# In[274]:


import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data=np.loadtxt("Book3.csv",delimiter = ";")
tab=[]
for i in range(np.size(data)):
    tab.append(int(round(data[i]/900)))

n=len(tab)

A=[]
B=[]

for j in range(0,n):
    if tab[j] not in A:
        A.append(tab[j])
        B.append(tab.count(tab[j]))        

plt.rcParams.update({'font.size': 15})

fig1, ax1 = plt.subplots() 
ax1 = plt.axes()
fig1 = plt.bar(A, B)

ax1.set_xlabel("Multiples of 900")  
ax1.set_ylabel("Number of lakes")  
ax1.set_title("REPARTITION OF LAKES SIZES (zoom)")

ax1.set_ylim(0,5100)
ax1.set_xlim(0,50)

X=np.log(B)
Y=np.log(A)

plt.show()

fig, (ax1, ax2) = plt.subplots(2, 1)
fig.suptitle('COMPARISON OF SIZE REPARTITION : number VS log')

ax1.plot(A,B)
ax1.set_ylabel('Numbers')

ax2.plot(X,Y)
ax2.set_xlabel('log')
ax2.set_ylabel('log')

#regression lin√©aire
z = np.polyfit(X, Y, 1)
p = np.poly1d(z)
plt.plot(X,p(X),"r--")

a=z[0]
b=z[1]

#calcul de R2
Ytilde=a*X+b
residuals=Ytilde-Y
R2=1-np.sum(residuals**2)/np.sum((Y-np.mean(Y))**2)

plt.show()
plt.show()

print("a=",a,"b=",b,"R**2=",R2)
```
